```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv('datasets_raw/mindmatv.csv')

# Excluindo colunas de teste
df_raw = df_raw.drop(range(0,5))
```

```python
#Extraindo as colunas do perfil

df_perfil = df_raw[['ID da resposta', 'Data de envio', 'Última página', 'Idioma inicial',
       'Semente', 'Data de início', 'Data da última ação', 'Nome Completo',
       'Idade', 'Sexo', 'Ano', 'Possui irmãos?',
       'Caso a resposta anterior seja positiva, quantos irmãos?',
       'Você costuma ler livros?', 'Você costuma ler quadrinhos?',
       'Você costuma ouvir músicas?', 'Você costuma assistir filmes e séries?',
       'Prefere ler ou ver séries/filmes?', 'Qual a sua matéria preferida?',
       'Rede social preferida', 'Horas por dia no celular',
       'Tipo de video preferido no YouTube',
       'Joga jogos eletrônicos? Se sim, onde?']]


# Retirando colunas irrelevantes para o perfil

df_perfil = df_perfil.drop(['ID da resposta', 'Data de envio', 'Última página', 'Idioma inicial',
       'Semente', 'Data de início', 'Data da última ação'], axis = 1)

# Resetando index e apagando o antigo

df_perfil = df_perfil.reset_index().drop('index', axis = 1)

# Dropando linhas com null
df_perfil = df_perfil.dropna()

# Renomear colunas

colunas = {'Nome Completo':'nome',
 'Idade':'idade', 
 'Sexo':'sexo', 
 'Ano':'ano', 
 'Possui irmãos?':'irmaos',
 'Caso a resposta anterior seja positiva, quantos irmãos?':'numirmaos',
 'Você costuma ler livros?':'lelivros',
 'Você costuma ler quadrinhos?':'lehqs',
 'Você costuma ouvir músicas?':'ouvemusicas',
 'Você costuma assistir filmes e séries?':'vefilmeserie',
 'Prefere ler ou ver séries/filmes?':'lerouverfilmeserie',
 'Qual a sua matéria preferida?':'materia',
 'Rede social preferida':'redesocial',
 'Horas por dia no celular': 'horascel',
 'Tipo de video preferido no YouTube': 'videoyt',
 'Joga jogos eletrônicos? Se sim, onde?': 'jogos'}

df_perfil = df_perfil.rename(columns = colunas)


# Tratando os dados das colunas 

# Passando os nomes para lowercase
df_perfil['nome'] = df_perfil['nome'].str.lower()


# Tratando coluna idade e convertendo para numerico

df_perfil['idade'] = df_perfil['idade'].str.split()
df_perfil['idade'] = df_perfil['idade'].str.join('')
df_perfil['idade'] = df_perfil['idade'].str.lower()
df_perfil['idade'] = df_perfil['idade'].str.replace('anos', '')
df_perfil['idade'] = df_perfil['idade'].str.replace('aninhos', '')
df_perfil['idade'] = df_perfil['idade'].str.replace('29/09/2010', '11')
df_perfil['idade'] = pd.to_numeric(df_perfil['idade'])

```
